````python
def prepare_data( data, fill_na=True, feature_selection=True, show_graph=False):
        """
        Preprocesses the dataset by filling missing values, encoding the target variable, 
        selecting features based on correlation, and scaling the feature set.
        
        Parameters:
        - data (pd.DataFrame): The input dataset to preprocess.
        - fill_na (bool): If True, fill missing values with the mean of each feature; otherwise, drop rows with missing values.
        - feature_selection (bool): If True, perform feature selection by removing highly correlated features (threshold: 0.9).
        - show_graph (bool): If True, displays graphs for missing values and correlation heatmap.
        
        Returns:
        - X (pd.DataFrame): The preprocessed feature matrix after scaling.
        - y (pd.Series): The encoded target variable.                  
        """
        from sklearn.preprocessing import MinMaxScaler
        from sklearn.preprocessing import LabelEncoder

        data = data.copy()

        y_feature = 'URL_Type_obf_Type'
        # Fill missing values with the mean of the respective columns if fill_na is True
        if (fill_na == True):
            data['avgpathtokenlen'] = data['avgpathtokenlen'].fillna(data['avgpathtokenlen'].mean())
            data['NumberRate_DirectoryName'] = data['NumberRate_DirectoryName'].fillna(data['NumberRate_DirectoryName'].mean())
            data['NumberRate_FileName'] = data['NumberRate_FileName'].fillna(data['NumberRate_FileName'].mean())
            data['NumberRate_Extension'] = data['NumberRate_Extension'].fillna(data['NumberRate_Extension'].mean())
            data['NumberRate_AfterPath'] = data['NumberRate_AfterPath'].fillna(data['NumberRate_AfterPath'].mean())
            data['Entropy_DirectoryName'] = data['Entropy_DirectoryName'].fillna(data['Entropy_DirectoryName'].mean())
            data['Entropy_Filename'] = data['Entropy_Filename'].fillna(data['Entropy_Filename'].mean())
            data['Entropy_Extension'] = data['Entropy_Extension'].fillna(data['Entropy_Extension'].mean())
            data['Entropy_Afterpath'] = data['Entropy_Afterpath'].fillna(data['Entropy_Afterpath'].mean())
        else:
            # Drop rows with missing values if fill_na is False
            data.dropna(axis='index', inplace=True)

        # Drop the 'argPathRatio' column since it does not affect the model's performance
        data = data.drop("argPathRatio", axis=1)
        # Plot missing values heatmap if show_graph is True
        if (show_graph == True):
            plt.figure(figsize=(10, 8))
            ax = plt.axes()
            sns.heatmap(data.isnull(), ax=ax, yticklabels=False, cbar=False, cmap="cividis")
            plt.show()

        # Encode the target variable using LabelEncoder
        le = LabelEncoder()
        data[y_feature] = le.fit_transform(data[y_feature]) 
        # Feature selection: Remove highly correlated features (correlation >= 0.9)
        if (feature_selection == True):
            corr = data.corr()        
            columns = np.full((corr.shape[0],), True, dtype=bool)
            for i in range(corr.shape[0]):
                for j in range(i+1, corr.shape[0]):
                    if corr.iloc[i,j] >= 0.9:
                        if columns[j]:
                            columns[j] = False
            selected_columns = data.columns[columns]
            data = data[selected_columns]            

        if (show_graph == True):
            corr = data.corr()
            plt.figure(figsize=(18,15))
            sns.heatmap(corr, annot=True, vmin=-1.0, cmap='cividis')
            plt.title('Correlation Heatmap')
            plt.show()
        
        # Scale the feature values using MinMaxScaler to bring all features to the same scale 
        scaler = MinMaxScaler()
        X = pd.DataFrame(scaler.fit_transform(data.loc[:, data.columns != y_feature]), columns=data.columns[:-1] )

        y = data[y_feature]
            
        return X, y
````

````python
def perform_anomaly_detection( X, y):
    """
    Detects and removes anomalies from the dataset using the Isolation Forest algorithm.
    
    Parameters:
    - X (pd.DataFrame): The input feature matrix.
    - y (pd.Series): The target variable.
    
    Returns:
    - X_new (pd.DataFrame): The feature matrix after removing detected anomalies.
    - y_new (pd.Series): The target variable after removing corresponding anomalies.
    """
    
    from sklearn.ensemble import IsolationForest

    # Instantiate the Isolation Forest model for anomaly detection
    iso_forest = IsolationForest(contamination=0.01, random_state=42).fit(X)
    
    # Predict anomalies: -1 indicates an anomaly, 1 indicates normal
    y_pred_iso_forest = iso_forest.predict(X)

    # Remove rows marked as anomalies (-1) from the dataset
    X_new = X[(y_pred_iso_forest != -1)]  # Keep rows that are not anomalies
    y_new = y[(y_pred_iso_forest != -1)]  # Keep corresponding target values

    # Print the new shapes of X and y after removing anomalies
    print('The shape after unsupervised anomaly detection:')
    print(f'X_new shape: {X_new.shape}')
    print(f'y_new shape: {y_new.shape}')
    
    return X_new, y_new
````

````python
def train_test_split( X, y, test_size, random_state, anomaly_detection=True):
        '''
        Splits the dataset into training and testing sets, and optionally performs anomaly detection 
        on both the training and testing data using Isolation Forest.

        Parameters:
        - X (pd.DataFrame): The feature matrix.
        - y (pd.Series): The target variable.
        - test_size (float): The proportion of the dataset to include in the test split (default is 0.2).
        - random_state (int): The seed used by the random number generator for reproducibility.
        - anomaly_detection (bool): If True, performs anomaly detection using Isolation Forest on the training and testing sets.
        
        Returns:
        - X_train (pd.DataFrame): Training feature matrix after optional anomaly detection.
        - X_test (pd.DataFrame): Testing feature matrix after optional anomaly detection.
        - y_train (pd.Series): Training target variable after optional anomaly detection.
        - y_test (pd.Series): Testing target variable after optional anomaly detection.
    
        '''
        from sklearn.model_selection import train_test_split

        # Split the dataset into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
                   
        print('The X_train, y_train shape:')
        print(X_train.shape)
        print(y_train.shape)
        if anomaly_detection == True:
            X_train, y_train = perform_anomaly_detection(X_train, y_train)

        print('The X_test, y_test shape:')
        print(X_test.shape)
        print(y_test.shape)

        
        if anomaly_detection == True:            
            X_test, y_test = perform_anomaly_detection(X_test, y_test)
                    
        return X_train, X_test, y_train, y_test  
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
````

````python
def initialize_models(random_state=42, max_iter=1000):
        """
        Initializes a dictionary of machine learning models to train.

        Parameters:
        - random_state (int): Seed for reproducibility (default is 42).
        - max_iter (int): Maximum number of iterations for the Logistic Regression model (default is 1000).

        Returns:
        - models (dict): A dictionary containing the initialized models with their respective names.
        """
        # Initialize an empty dictionary to hold models
        models = {}
        models['DecisionTreeClassifier'] = DecisionTreeClassifier(random_state=random_state)    
        models['LogisticRegression'] = LogisticRegression(random_state=random_state, max_iter=max_iter)
        models['RandomForestClassifier'] = RandomForestClassifier(random_state=random_state)

        return models
````

````python
def define_model_parameters(use_best_params=False):
        '''
        Defines the hyperparameters for DecisionTreeClassifier and LogisticRegression models.

        Parameters:
        - use_best_params (bool): If True, returns pre-defined best hyperparameters; 
                                if False, returns a wider range of parameters for tuning.

        Returns:
        - params (dict): A dictionary containing the hyperparameters for both models. 
        '''

        # Define hyperparameter dictionaries for DecisionTreeClassifier and LogisticRegression
        params_tree = {}
        params_logistic = {}
        params_random = {}

        if use_best_params == False:
            # Parameters to search over during tuning
            params_tree['criterion'] = ['gini', 'entropy']  # Criterion for splitting
            params_tree['max_depth'] = [1, 3, 5, 10]        # Different max depths for the tree
            params_tree['min_samples_split'] = [5, 10]      # Minimum samples to split a node
            params_tree['min_samples_leaf'] = [5, 10]       # Minimum samples in a leaf node

            params_logistic['solver'] = ['liblinear']       # Solver used for optimization
            params_logistic['penalty'] = ['l2']             # Regularization type
            params_logistic['C'] = [100, 10, 1.0, 0.1]      # Inverse regularization strength
            
            params_random['criterion'] = ['gini', 'entropy'] # Criterion for splitting
            params_random['n_estimators'] = [100, 150, 200] # Number of trees in the forest. Higher values typically increase accuracy but may increase computation time.
            params_random['max_depth'] = [1, 3, 5, 10] # Different max depths for the tree
            params_random['min_samples_split'] = [5, 10] # Minimum samples to split a node
            params_random['min_samples_leaf'] = [5, 10]   # Minimum samples in a leaf node              

        else:
            # Pre-selected best parameters
            params_tree['criterion'] = ['entropy']
            params_tree['max_depth'] = [10]         
            params_tree['min_samples_leaf'] = [5]  
            params_tree['min_samples_split'] = [5]       

            params_logistic['solver'] = ['liblinear']
            params_logistic['penalty'] = ['l2']
            params_logistic['C'] = [100]

            params_random['criterion'] = ['entropy']
            params_random['n_estimators'] = [200]
            params_random['max_depth'] = [10]
            params_random['min_samples_split'] = [5] 
            params_random['min_samples_leaf'] = [5]  

         # Combine both models' hyperparameters into one dictionary
        params = {
            'DecisionTreeClassifier': params_tree,
            'LogisticRegression': params_logistic,
            "RandomForestClassifier" : params_random
        }

        return params
````

````python
models = initialize_models()
params = define_model_parameters()
keys=models.keys()
grid_searches ={}
````

````python
def fit(X, y, scoring=None, n_jobs=3, refit=True, cv=5, verbose=1, return_train_score=True):
        """
        Performs Grid Search with Cross-Validation for multiple models and their hyperparameters.

        Parameters:
        - X (pd.DataFrame): Feature matrix for training.
        - y (pd.Series): Target variable for training.
        - scoring (str or callable): A single string or callable to evaluate the predictions on the test set.
        - n_jobs (int): Number of jobs to run in parallel. (default: 3)
        - refit (bool): Refit an estimator using the best found parameters on the whole dataset. (default: True)
        - cv (int): Number of cross-validation folds. (default: 5)
        - verbose (int): Controls the verbosity: the higher, the more messages. (default: 1)
        - return_train_score (bool): If False, the `cv_results_` attribute will not include training scores. (default: True)
        """
        for key in keys:
            print('---------------------------------------------------------------------------')
            print(f'Training model: {key}')
            
            model = models[key]
            param = params[key]
            
            grid_search = GridSearchCV(model, param, cv=cv, n_jobs=n_jobs,
                verbose=verbose, scoring=scoring, refit=refit,
                return_train_score=return_train_score)
            
            # Perform the grid search on the data
            grid_search.fit(X, y)
            
            grid_searches[key] = grid_search
            
            # If refit is True, print the best parameters and score
            if (refit==True): 
                print('best_params_: ', grid_search.best_params_)
                print('best_score_: ', grid_search.best_score_)
````

````python
fit(X_train,y_train,scoring='accuracy')
````

````python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

def evaluate_models(grid_searches, X_test, y_test):
    """
    Evaluates models using accuracy and weighted F1 score, and prints the classification report.

    Parameters:
    - grid_searches (dict): Dictionary of trained GridSearchCV objects.
    - X_test (pd.DataFrame): Test feature data.
    - y_test (pd.Series): True labels for the test data.

    Returns:
    - results (list): A list containing model names, accuracy, and F1 scores.
    """
    
    # List to store evaluation results
    results = []

    # Iterate over each model and evaluate performance
    for key, grid_search in grid_searches.items():
        print('---------------------------------------------------------------------------')
        print(f'Evaluating model: {key}')
        
        # Retrieve the best model from GridSearchCV
        best_model = grid_search.best_estimator_
        
        # Predict on the test set
        y_pred = best_model.predict(X_test)

        # Calculate accuracy and F1 score
        acc = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred, average='weighted')

        # Print the classification report for the model
        print(f'Classification Report for {key}:\n', classification_report(y_test, y_pred))
        
        # Append the results to the list
        results.append([key, acc, f1])
    
    return results

# Call the function and store the results
results = evaluate_models(grid_searches, X_test, y_test)

````

````python
df_test = pd.DataFrame(results, columns=['Model', 'Accuracy', 'F1-score'])
df_test = df_test.sort_values(by='Accuracy', ascending=False)
df_test
````
